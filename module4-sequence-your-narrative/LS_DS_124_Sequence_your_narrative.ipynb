## Merge data

https://github.com/pandas-dev/pandas/blob/master/doc/cheatsheet/Pandas_Cheat_Sheet.pdf

```python
df = pd.merge(income, lifespan)
df.shape
```

```python
df = pd.merge(df, population)
df.shape
```

```python
entities_var = ['country', 'name', 'world_6region']

# entities['world_6region'].value_counts()
# entities['world_4region'].value_counts()
```

```python
entities = entities[entities_var]

print(entities.shape)
entities.head()
```

```python
merged = pd.merge(df, entities, left_on='geo', right_on='country')

print(df.shape)
df.head()
```

```python
merged = merged.drop(['geo','country'], axis='columns')
```

```python
merged = merged.rename( columns = {
    'time': 'year',
    'income_per_person_gdppercapita_ppp_inflation_adjusted': 'income',
    'life_expectancy_years':'lifespan',
    'population_total': 'population',
    'name':'country',
    'world_6region': 'region'
})


merged.head()
```

## Plot visualization

```python
import seaborn as sns

now = merged[merged['year'] == 2018]

sns.relplot(x='income', y='lifespan', hue='region', size='population', data=now);
```

```python
# now[now.income > 80000].sort_values(by='income')
qatar_now = now[(now.income > 80000) & (now.country == 'Qatar')].sort_values(by='income')
qatar_now
```

```python
sns.relplot(x='income', y='lifespan', hue='region', size='population', data=now);

plt.text(x=qatar_now.income , y=qatar_now.lifespan + 1, s='Qatar')
```

```python
years = [1818, 1918, 2018]
centuries = merged[merged.year.isin(years)]

sns.relplot(x='income', y='lifespan', hue='region', size='population', col='year', data=centuries);
plt.xscale('log')

```

```python
new_years = [1818, 1938, 1958, 1978, 1998, 2018]
time_w_new_years = merged[merged.year.isin(new_years)]

sns.relplot(x='income', y='lifespan', hue='region', size='population', col='year', data=time_w_new_years);
```

```python
for year in years: 
  sns.relplot(x='income', y='lifespan', hue='region', 
              size='population', data=merged[merged.year == year])
  
  plt.xscale('log')
  plt.xlim((150,150000))
  plt.ylim((0,90))
  plt.title(year)
  plt.axhline(y=50, color='grey')
```

```python
year = 1862 #@param {type: 'slider', min:1800, max:2018, step:1}

sns.relplot(x='income', y='lifespan', hue='region', 
            size='population', data=merged[merged.year == year])

plt.xscale('log')
plt.xlim((150,150000))
plt.ylim((0,90))
plt.title(year)
```

```python
from google.colab import widgets
tb = widgets.TabBar([str(year) for year in years])
for tab, year in zip(tb,years):
  sns.relplot(x='income', y='lifespan', hue='region', 
            size='population', data=merged[merged.year == year])
  
  plt.xscale('log')
  plt.xlim((150,150000))
  plt.ylim((0,90))
```
